# Clustering Analysis and Interpretation: Dow Jones Index Dataset

This notebook performs a comprehensive clustering analysis on the Dow Jones Index dataset, including:

- Clustering with multiple algorithms
- Statistical testing for cluster differences

- Feature importance and SHAP analysis
- Comprehensive visualizations

## 1. Load and Explore the Dataset

We'll start by loading the dataset and examining its structure, checking for missing values, and computing basic statistics.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Fetch dataset
print("Loading Dow Jones Index dataset...")
dow_jones_index = fetch_ucirepo(id=312)

# Extract features and target
X = dow_jones_index.data.features
y = dow_jones_index.data.targets

# Combine for easier analysis
df = pd.concat([X, y], axis=1)

print(f"\nDataset shape: {df.shape}")
print(f"Number of samples: {len(df)}")
print(f"Number of features: {len(X.columns)}")
```

```python
# Display first few rows
print("First 5 rows of the dataset:")
display(df.head())

# Display data types
print("\nData types:")
print(df.dtypes)
```

```python
# Check for missing values
print("Missing values per column:")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0] if missing_values.sum() > 0 else "No missing values found!")

# Basic statistics for numerical columns
print("\n" + "="*80)
print("Basic Statistics for Numerical Features:")
print("="*80)
display(df.describe())
```

```python
# Examine categorical features
print("Unique values in categorical columns:")
print(f"\nNumber of unique stocks: {df['stock'].nunique()}")
print(f"Stocks: {df['stock'].unique()}")
print(f"\nQuarters: {df['quarter'].unique()}")
print(f"\nDate range: {df['date'].min()} to {df['date'].max()}")
```

## 2. Data Preprocessing and Feature Engineering

Convert categorical price features to numeric values, handle date features, encode categorical variables, and prepare the feature matrix for clustering.

```python
# Create a copy for preprocessing
df_processed = df.copy()

# Convert price columns from string to float (remove '$' if present)
price_columns = ['open', 'high', 'low', 'close', 'next_weeks_open', 'next_weeks_close']

for col in price_columns:
    if df_processed[col].dtype == 'object':
        # Remove '$' and convert to float
        df_processed[col] = df_processed[col].astype(str).str.replace('$', '').astype(float)
        print(f"Converted {col} to numeric")

print("\nPrice columns after conversion:")
print(df_processed[price_columns].dtypes)
```

```python
# Extract date features
df_processed['date'] = pd.to_datetime(df_processed['date'])
df_processed['year'] = df_processed['date'].dt.year
df_processed['month'] = df_processed['date'].dt.month
df_processed['day_of_week'] = df_processed['date'].dt.dayofweek
df_processed['week_of_year'] = df_processed['date'].dt.isocalendar().week

print("Date features extracted:")
print(df_processed[['date', 'year', 'month', 'day_of_week', 'week_of_year']].head())
```

```python
# Encode stock names (categorical variable)
from sklearn.preprocessing import LabelEncoder

le_stock = LabelEncoder()
df_processed['stock_encoded'] = le_stock.fit_transform(df_processed['stock'])

# Create a mapping for reference
stock_mapping = dict(zip(le_stock.classes_, range(len(le_stock.classes_))))
print("Stock encoding mapping:")
for name, code in stock_mapping.items():
    print(f"  {code}: {name}")
```

```python
# Select features for clustering
# Exclude non-informative columns: original date, stock name
clustering_features = [
    'quarter', 'stock_encoded', 'year', 'month', 'week_of_year',
    'open', 'high', 'low', 'close', 'volume',
    'percent_change_price', 'percent_change_volume_over_last_wk',
    'previous_weeks_volume', 'next_weeks_open', 'next_weeks_close',
    'days_to_next_dividend', 'percent_return_next_dividend'
]

# Create feature matrix for clustering
X_cluster = df_processed[clustering_features].copy()

print(f"Feature matrix shape: {X_cluster.shape}")
print(f"\nFeatures for clustering:")
for i, feat in enumerate(clustering_features, 1):
    print(f"  {i}. {feat}")

# Check for any remaining missing values
if X_cluster.isnull().sum().sum() > 0:
    print("\nWarning: Missing values detected. Filling with median...")
    X_cluster = X_cluster.fillna(X_cluster.median())
else:
    print("\nNo missing values in clustering features!")
```

```python
# Standardize features for clustering
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

print("Features have been standardized (mean=0, std=1)")
print(f"Scaled data shape: {X_scaled.shape}")
print(f"\nSample statistics after scaling:")
print(f"  Mean: {X_scaled.mean():.6f}")
print(f"  Std: {X_scaled.std():.6f}")
```

## 3. Perform Clustering Analysis

Apply K-Means and Hierarchical Clustering, determine optimal number of clusters using elbow method and silhouette scores.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Test different numbers of clusters
k_range = range(2, 11)
inertias = []
silhouette_scores = []
davies_bouldin_scores = []
calinski_harabasz_scores = []

print("Evaluating different numbers of clusters...")
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, labels))
    davies_bouldin_scores.append(davies_bouldin_score(X_scaled, labels))
    calinski_harabasz_scores.append(calinski_harabasz_score(X_scaled, labels))
    
    print(f"k={k}: Silhouette={silhouette_scores[-1]:.3f}, Davies-Bouldin={davies_bouldin_scores[-1]:.3f}")
```

```python
# Visualize clustering metrics
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Elbow plot
axes[0, 0].plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
axes[0, 0].set_xlabel('Number of Clusters (k)', fontsize=12)
axes[0, 0].set_ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=12)
axes[0, 0].set_title('Elbow Method', fontsize=14, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)

# Silhouette score
axes[0, 1].plot(k_range, silhouette_scores, 'go-', linewidth=2, markersize=8)
axes[0, 1].set_xlabel('Number of Clusters (k)', fontsize=12)
axes[0, 1].set_ylabel('Silhouette Score', fontsize=12)
axes[0, 1].set_title('Silhouette Score (Higher is Better)', fontsize=14, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)

# Davies-Bouldin score
axes[1, 0].plot(k_range, davies_bouldin_scores, 'ro-', linewidth=2, markersize=8)
axes[1, 0].set_xlabel('Number of Clusters (k)', fontsize=12)
axes[1, 0].set_ylabel('Davies-Bouldin Score', fontsize=12)
axes[1, 0].set_title('Davies-Bouldin Score (Lower is Better)', fontsize=14, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3)

# Calinski-Harabasz score
axes[1, 1].plot(k_range, calinski_harabasz_scores, 'mo-', linewidth=2, markersize=8)
axes[1, 1].set_xlabel('Number of Clusters (k)', fontsize=12)
axes[1, 1].set_ylabel('Calinski-Harabasz Score', fontsize=12)
axes[1, 1].set_title('Calinski-Harabasz Score (Higher is Better)', fontsize=14, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Recommend optimal k
optimal_k_silhouette = k_range[np.argmax(silhouette_scores)]
print(f"\nRecommended number of clusters based on Silhouette Score: {optimal_k_silhouette}")
```

```python
# Perform final clustering with optimal k
optimal_k = 7  # You can adjust this based on the metrics above

print(f"Performing K-Means clustering with k={optimal_k}...")
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels_kmeans = kmeans_final.fit_predict(X_scaled)

# Add cluster labels to dataframe
df_processed['cluster_kmeans'] = cluster_labels_kmeans

print(f"\nCluster distribution (K-Means):")
print(df_processed['cluster_kmeans'].value_counts().sort_index())
print(f"\nCluster proportions:")
print(df_processed['cluster_kmeans'].value_counts(normalize=True).sort_index())
```

```python
# Also try Hierarchical Clustering for comparison
from sklearn.cluster import AgglomerativeClustering

print(f"Performing Hierarchical Clustering with {optimal_k} clusters...")
hierarchical = AgglomerativeClustering(n_clusters=optimal_k, linkage='ward')
cluster_labels_hier = hierarchical.fit_predict(X_scaled)

df_processed['cluster_hierarchical'] = cluster_labels_hier

print(f"\nCluster distribution (Hierarchical):")
print(df_processed['cluster_hierarchical'].value_counts().sort_index())

# We'll use K-Means clusters for the rest of the analysis
df_processed['cluster'] = df_processed['cluster_kmeans']
print(f"\nUsing K-Means clustering results for further analysis.")
```

## 5. Statistical Testing for Cluster Differences

Perform statistical tests to determine significant differences between clusters.

```python
from scipy import stats
from scipy.stats import f_oneway, kruskal, chi2_contingency, fisher_exact
from statsmodels.stats.multitest import multipletests

# Function to calculate Cohen's d effect size
def cohens_d(group1, group2):
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2))
    return (np.mean(group1) - np.mean(group2)) / pooled_std

print("="*80)
print("STATISTICAL TESTS FOR CONTINUOUS FEATURES")
print("="*80)
```

```python
# Perform ANOVA and Kruskal-Wallis tests for continuous features
test_results = []

for feature in continuous_features:
    # Group data by cluster
    groups = [df_processed[df_processed['cluster'] == c][feature].values 
              for c in range(optimal_k)]
    
    # ANOVA test (parametric)
    f_stat, p_anova = f_oneway(*groups)
    
    # Kruskal-Wallis test (non-parametric)
    h_stat, p_kruskal = kruskal(*groups)
    
    # Calculate effect size (eta-squared for ANOVA)
    grand_mean = df_processed[feature].mean()
    ss_between = sum([len(g) * (g.mean() - grand_mean)**2 for g in groups])
    ss_total = sum([(x - grand_mean)**2 for g in groups for x in g])
    eta_squared = ss_between / ss_total if ss_total > 0 else 0
    
    test_results.append({
        'feature': feature,
        'f_statistic': f_stat,
        'p_value_anova': p_anova,
        'h_statistic': h_stat,
        'p_value_kruskal': p_kruskal,
        'eta_squared': eta_squared
    })

# Create results dataframe
test_results_df = pd.DataFrame(test_results)
test_results_df = test_results_df.round(6)

# Apply Benjamini-Hochberg correction
test_results_df['p_adjusted_anova'] = multipletests(test_results_df['p_value_anova'], 
                                                      method='fdr_bh')[1]
test_results_df['p_adjusted_kruskal'] = multipletests(test_results_df['p_value_kruskal'], 
                                                        method='fdr_bh')[1]

# Mark significant results
test_results_df['significant_anova'] = test_results_df['p_adjusted_anova'] < 0.05
test_results_df['significant_kruskal'] = test_results_df['p_adjusted_kruskal'] < 0.05

print("\nStatistical Test Results for Continuous Features:")
display(test_results_df)
```

```python
# Calculate pairwise Cohen's d for significant features
print("\n" + "="*80)
print("PAIRWISE EFFECT SIZES (Cohen's d) FOR SIGNIFICANT FEATURES")
print("="*80)

significant_features = test_results_df[test_results_df['significant_anova']]['feature'].values

for feature in significant_features[:5]:  # Show top 5 for brevity
    print(f"\n{feature}:")
    effect_sizes = []
    
    for i in range(optimal_k):
        for j in range(i+1, optimal_k):
            group_i = df_processed[df_processed['cluster'] == i][feature].values
            group_j = df_processed[df_processed['cluster'] == j][feature].values
            
            d = cohens_d(group_i, group_j)
            effect_sizes.append({
                'Cluster_i': i,
                'Cluster_j': j,
                'Cohens_d': round(d, 3),
                'Effect_Size': 'Small' if abs(d) < 0.5 else 'Medium' if abs(d) < 0.8 else 'Large'
            })
    
    effect_df = pd.DataFrame(effect_sizes)
    display(effect_df)
```

```python
# Chi-square tests for categorical features
print("\n" + "="*80)
print("CHI-SQUARE TESTS FOR CATEGORICAL FEATURES")
print("="*80)

categorical_test_results = []

# Test for stock
contingency_stock = pd.crosstab(df_processed['cluster'], df_processed['stock'])
chi2_stock, p_stock, dof_stock, expected_stock = chi2_contingency(contingency_stock)

categorical_test_results.append({
    'feature': 'stock',
    'chi2_statistic': chi2_stock,
    'p_value': p_stock,
    'degrees_of_freedom': dof_stock
})

# Test for quarter
contingency_quarter = pd.crosstab(df_processed['cluster'], df_processed['quarter'])
chi2_quarter, p_quarter, dof_quarter, expected_quarter = chi2_contingency(contingency_quarter)

categorical_test_results.append({
    'feature': 'quarter',
    'chi2_statistic': chi2_quarter,
    'p_value': p_quarter,
    'degrees_of_freedom': dof_quarter
})

categorical_results_df = pd.DataFrame(categorical_test_results)
categorical_results_df['p_adjusted'] = multipletests(categorical_results_df['p_value'], 
                                                       method='fdr_bh')[1]
categorical_results_df['significant'] = categorical_results_df['p_adjusted'] < 0.05

print("\nChi-Square Test Results:")
display(categorical_results_df)
```

## 6. Multivariate Analysis with Random Forest

Train a Random Forest classifier to predict cluster labels and evaluate model performance.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Prepare data for Random Forest
X_rf = X_cluster.values
y_rf = df_processed['cluster'].values

print(f"Training Random Forest to predict cluster labels...")
print(f"Features: {X_rf.shape[1]}")
print(f"Samples: {X_rf.shape[0]}")
print(f"Classes: {len(np.unique(y_rf))}")
```

```python
# Train Random Forest with cross-validation
rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)

# Perform cross-validation
cv_scores = cross_val_score(rf_model, X_rf, y_rf, cv=5, scoring='accuracy')

print("\nCross-Validation Results:")
print(f"  Accuracy scores: {cv_scores}")
print(f"  Mean accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
```

```python
# Train on full dataset for feature importance analysis
rf_model.fit(X_rf, y_rf)

# Make predictions
y_pred = rf_model.predict(X_rf)

# Evaluate model
print("\nModel Performance on Training Data:")
print(f"Accuracy: {accuracy_score(y_rf, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_rf, y_pred))
```

```python
# Confusion matrix
cm = confusion_matrix(y_rf, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=[f'Cluster {i}' for i in range(optimal_k)],
            yticklabels=[f'Cluster {i}' for i in range(optimal_k)])
plt.title('Confusion Matrix - Random Forest Cluster Prediction', fontsize=14, fontweight='bold')
plt.ylabel('True Cluster', fontsize=12)
plt.xlabel('Predicted Cluster', fontsize=12)
plt.tight_layout()
plt.show()

print(f"\nThe high accuracy indicates that clusters are well-separated and features are informative.")
```

## Summary and Interpretation

Let's summarize the key findings from our clustering analysis.

```python
# Generate comprehensive cluster interpretation
print("="*80)
print("CLUSTER INTERPRETATION SUMMARY")
print("="*80)

for cluster_idx in range(optimal_k):
    print(f"\n{'='*80}")
    print(f"CLUSTER {cluster_idx}")
    print(f"{'='*80}")
    
    cluster_data = df_processed[df_processed['cluster'] == cluster_idx]
    
    print(f"\nSize: {len(cluster_data)} samples ({len(cluster_data)/len(df_processed)*100:.1f}%)")
    
    print("\nKey Characteristics (Top Features):")
    
    # Get top distinguishing features for this cluster
    cluster_means = df_processed[df_processed['cluster'] == cluster_idx][continuous_features].mean()
    overall_means = df_processed[continuous_features].mean()
    differences = ((cluster_means - overall_means) / overall_means * 100).sort_values(ascending=False)
    
    print("\n  Features Above Average:")
    for feat, diff in differences.head(5).items():
        print(f"    - {feat}: {diff:+.1f}% from overall mean")
    
    print("\n  Features Below Average:")
    for feat, diff in differences.tail(5).items():
        print(f"    - {feat}: {diff:+.1f}% from overall mean")
    
    print("\n  Most Common Stocks:")
    top_stocks = cluster_data['stock'].value_counts().head(3)
    for stock, count in top_stocks.items():
        print(f"    - {stock}: {count} samples ({count/len(cluster_data)*100:.1f}%)")
    
    print("\n  Quarter Distribution:")
    quarter_dist = cluster_data['quarter'].value_counts().sort_index()
    for quarter, count in quarter_dist.items():
        print(f"    - Q{quarter}: {count} samples ({count/len(cluster_data)*100:.1f}%)")
```
